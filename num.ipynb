x+3y=11
3x+6y =4

```python
import numpy as np
coff=np.array([[1,3],[3,6]])
const = np.array([11,4])
x = np.linalg.solve(coff,const)
print(x)
```

x+2y+z=-2
4x+9y+2z=7
5x-8y-9z=12

```python

coff=np.array([[1,2,1],[4,9,2],[5,-8,-9]])
const = np.array([-2,7,12])
x1 = np.linalg.solve(coff,const)
print(x1)
```
